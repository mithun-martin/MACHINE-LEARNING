```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris = load_iris()
X = iris.data[:, 0]
# Extracts only the first feature (column) of the dataset — sepal length — and stores it in X.
# iris.data is a 2D array; [:, 0] means “all rows, first column.”

relu = np.maximum(0, X)
# Calculates ReLU activation: values less than 0 become 0, others stay the same.
leaky_relu = np.where(X > 0, X, 0.1 * X)
# Calculates Leaky ReLU: positive values stay the same, negative values are multiplied by 0.1 instead of being set to 0.
sigmoid = 1 / (1 + np.exp(-X))
# Calculates Sigmoid activation: maps all numbers to a range between 0 and 1.
tanh = np.tanh(X)
# Calculates Tanh activation: maps all numbers to a range between -1 and 1.

plt.figure(figsize=(8,5))
plt.plot(X, relu, 'o', label='ReLU')
plt.plot(X, leaky_relu, 'o', label='Leaky ReLU')
plt.plot(X, sigmoid, 'o', label='Sigmoid')
plt.plot(X, tanh, 'o', label='Tanh')

plt.title("Activation Functions on Iris Feature (Sepal Length)")
plt.xlabel("Original Value")
plt.ylabel("Transformed Value")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data, img_as_float



image = img_as_float(data.astronaut())



# Activation functions
def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.1):
    return np.where(x > 0, x, alpha * x)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh_act(x):
    return np.tanh(x)

relu_img = relu(image - 0.5)
leaky_relu_img = leaky_relu(image - 0.5)
sigmoid_img = sigmoid(image - 0.5)
tanh_img = tanh_act(image - 0.5)

fig, axs = plt.subplots(1, 5, figsize=(15, 5))
axs[0].imshow(image, cmap='gray')
axs[0].set_title("Original")
axs[0].axis('off')

axs[1].imshow(relu_img,cmap='gray')
axs[1].set_title("ReLU")
axs[1].axis('off')

axs[2].imshow(leaky_relu_img, cmap='gray')
axs[2].set_title("Leaky ReLU")
axs[2].axis('off')

axs[3].imshow(sigmoid_img, cmap='gray')
axs[3].set_title("Sigmoid")
axs[3].axis('off')

axs[4].imshow(tanh_img, cmap='gray')
axs[4].set_title("Tanh")
axs[4].axis('off')

plt.tight_layout()
plt.show()
```

```python
# We subtract 0.5 from the image because pixel values in img_as_float() are scaled to the range [0, 1], and:

# 0.5 is the midpoint.

# Subtracting 0.5 shifts the range from [0, 1] → [-0.5, 0.5].

# Why this matters for activation functions:
# ReLU: If all pixel values are already positive (0–1), negative-side behavior isn’t visible. Shifting down introduces negatives so you can see how the function handles them.

# Leaky ReLU: Shows the “small slope” for negative values.

# Sigmoid/Tanh: Centering around 0 lets these functions map values symmetrically (especially Tanh which outputs -1 to 1).

# 🔍 Example without subtraction (0–1 range):

# ReLU and Leaky ReLU would look almost identical.

# Tanh would give only positive outputs.
```
