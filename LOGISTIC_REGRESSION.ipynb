```python
#✅ Yes, you must handle NaNs before fitting the model.
#Either drop or fill them.
#Why? Because sklearn’s LogisticRegression implementation requires clean, numeric input
```

```python

#Dropping Rows with Nan in all provided columns
data = data.dropna(subset=["sex","pclass","survived"])
```

```python

#Again Finding  number of Nan or missing values in all columns
print(data.isna().sum(axis = 0))
```

```python
#Preprocessing: Label Encoding
#Since in sex column we are not having a numerical vallue, just having male and female.
#Since, model doesn't accept text, wo we will transform text into binary labels.

```

```python
data["sex"] = data["sex"].map({"male":1,"female":0})
print(data["sex"])
```

```python
#NOW EXTRACTING FEAUTRE
#seperating feautre attributes from taret attribute
```

```python
feature = data[["sex",'age',"pclass"]]
target = data[["survived"]]
```

```python
#📌 Logic is:
#If too many missing values in an important numeric column (like age here → 263 missing out of ~1300 rows)
#👉 Dropping would waste a lot of data
#👉 So better to impute (fill) using mean/median/mode depending on the data type and distribution.
```

```python
#Preprocessing: Imputation | Missing Values in Age
#Now since we are having so many missing values in our column of age we will Impute the entire column
```

```python
from sklearn.impute import SimpleImputer

# ✅ Creating an imputer object
# missing_values = np.nan → this tells the imputer which values to consider as missing (NaN in this case)
# strategy='mean' → this tells it to replace missing values with the mean of that particular column
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')

# ✅ Fitting the imputer on our features data and transforming it
# fit_transform() does 2 things:
# 1️⃣ Calculates the mean of each column (fit)
# 2️⃣ Replaces all the NaN values with the respective column mean (transform)
feature = imputer.fit_transform(feature)
```

```python
feature
```

```python
#test_size=0.2, random_state=0
```

```python

feature_train, feature_test, target_train, target_test = train_test_split(feature,target)
```

```python

model = LogisticRegression()
model.fit(feature_train, target_train)
predictions = model.predict(feature_test)
```
