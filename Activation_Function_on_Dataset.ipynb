```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

# Load Iris dataset
iris = load_iris()
X = iris.data  # shape (150, 4)

# We'll use only first two features for visualization
X_2d = X[:, :2]
# Why first 2 features?
# Because when we want to plot data on a 2D graph, we need exactly two dimensions  one for the x-axis and one for the y-axis.
# The Iris dataset has 4 features (4 columns), but we cant plot 4D data directly on a simple 2D plot.
# So, we just pick the first 2 columns (features) to visualize the data points in a 2D scatter plot.
# Each point on the plot will be:
# x-axis = first feature
# y-axis = second feature

# Apply activation functions
sigmoid_out = sigmoid(X_2d)
tanh_out = tanh(X_2d)
relu_out = relu(X_2d)
leaky_relu_out = leaky_relu(X_2d)

# Plot original and transformed data
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.scatter(X_2d[:, 0], X_2d[:, 1], c='blue', alpha=0.5)
plt.title("Original Data (2 features)")

plt.subplot(2, 3, 2)
plt.scatter(sigmoid_out[:, 0], sigmoid_out[:, 1], c='green', alpha=0.5)
plt.title("After Sigmoid")

plt.subplot(2, 3, 3)
plt.scatter(tanh_out[:, 0], tanh_out[:, 1], c='red', alpha=0.5)
plt.title("After Tanh")

plt.subplot(2, 3, 4)
plt.scatter(relu_out[:, 0], relu_out[:, 1], c='purple', alpha=0.5)
plt.title("After ReLU")

plt.subplot(2, 3, 5)
plt.scatter(leaky_relu_out[:, 0], leaky_relu_out[:, 1], c='orange', alpha=0.5)
plt.title("After Leaky ReLU")

plt.tight_layout()
plt.show()
```
