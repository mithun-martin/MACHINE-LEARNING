```python

#importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
import math
```

```python
df = pd.read_csv("/content/Drugs_Performance.csv")
df.head()
```

```python
X = df[['Age']].values
y = df[['Drugs_Perf']].values
print(X)
print(y)
```

```python

plt.plot(X,y)
plt.xlabel("Age")
plt.ylabel("Drugs_Perf")
plt.show()
```

```python

X_train, X_test, Y_train, Y_test = train_test_split(X,y, test_size=0.2)
```

```python
#now before trainng teh model 4 polynal regression
```

```python
#we need to preprocess our data for the nth polynomial feutres
```

```python
poly = PolynomialFeatures(degree = 2)
#This creates a PolynomialFeatures object from scikit-learn.
#degree=2 means it’ll create:
#x,x²,and a constant term (1)


X_poly = poly.fit_transform(X_train)
#It takes your X_train values and transforms it into polynomial terms

#first we create a linear regression model
model = LinearRegression()
model.fit(X_poly, Y_train)

#now before giving ypred=model.predict(xtest) we need to transform test data
#also into the format of xtrain data

X_test_poly = poly.transform(X_test)
# Transform the test data into polynomial features using the same transformation learned from training data
# We use .transform() (not .fit_transform()) because the feature mapping (degree and term arrangement)
# was already decided when we called .fit_transform() on the training data.
# This ensures that test data has the same structure (columns for 1, x, x², etc.) as the training data.


```

```python
y_pred = model.predict(X_test_poly)
```
