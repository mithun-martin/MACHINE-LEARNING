```python
# üìå K-Means Clustering Algorithm üìå
# It is an Unsupervised Machine Learning algorithm used to group unlabeled data into clusters
# based on similarity between data points (commonly via Euclidean distance).

# üìå Example Use Case:
# An online store uses K-Means to group customers based on purchase frequency & spending,
# creating groups like 'Budget Shoppers', 'Frequent Buyers', and 'Big Spenders'.

# üìå How K-Means Works:
# 1Ô∏è‚É£ Randomly pick 'k' points as initial cluster centroids (starting positions)
# 2Ô∏è‚É£ Assign each data point to the nearest centroid ‚Äî this forms the clusters
# 3Ô∏è‚É£ Update the centroids by taking the mean position of all points in a cluster
# 4Ô∏è‚É£ Repeat steps 2 and 3 until the centroids no longer change (convergence)
```

```python
# üì¶ Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
# üìå Generate synthetic dataset with 3 clusters (for demo)

X,y = make_blobs(n_samples=500, n_features=2, centers=3)






#X ‚Üí input data points (features)
#A 2D NumPy array of shape (500, 2) ‚Üí 500 points, each with 2 features (like x and y coordinates for 2D plotting)
#y ‚Üí actual cluster labels

#üìå What‚Äôs happening here:
#This line is generating a synthetic (fake) dataset just for demo/testing clustering algorithms like K-Means. It creates: A bunch of data points (X)
#And their true cluster labels (y), but those labels are not used during clustering ‚Äî
#because clustering is unsupervised (we don‚Äôt tell the algorithm the labels; it has to figure out clusters by itself).

#n_samples=500 ‚Üí generate 500 total data points
#centers=3 ‚Üí create 3 natural groups (clusters) in the data
#n_features=2 ‚Üí each data point has 2 features (x and y coordinates), so we can visualize them in a 2D scatter plot
```

```python
plt.scatter(X[:, 0],X[:, 1])
# üìå Plotting the original dataset points before clustering
# X[:, 0] ‚Üí selects all values from the first feature (column 0) to plot on x-axis
# X[:, 1] ‚Üí selects all values from the second feature (column 1) to plot on y-axis
# plt.scatter() ‚Üí plots a scatter plot of those data points

plt.title("original data")
```

```python
k = 3
```

```python
# üìå Randomly initialize cluster centers (centroids)
np.random.seed(42)
centroids = X[np.random.choice( range(len(X)) , k, replace=False ),]

#np.random.choice()	NumPy‚Äôs function to randomly pick values from a given list or range.

#range(len(X)) Creates a sequence of numbers from 0 to len(X) - 1 (i.e., index numbers of all data points in X).
#If your dataset X has 100 rows, range(len(X)) creates [0,1,2,...,99]
#If you skip it and give just k ‚Äî like np.random.choice(k, k, ...) ‚Äî it only picks from [0,1,2,...,k-1],
#meaning your centroids would come from just the first k rows of X, which is both biased and incorrect for clustering.

# k The number of centroids (clusters) you want to initialize.

#replace = False Ensures that no index is picked more than once ‚Äî meaning you‚Äôll get k unique data points as centroids, no repetition.
```

```python
#1 we need to calc dis bw the datapoint to cluseer so we dif ed func
```

```python
def euclidean_distance(a, b):
  return np.sqrt(np.sum(  (a-b)**2 ))
```

```python

# üìå K-Means Main Iteration Loop
for iterations in range(10):                # 10 times run tilll taht muc we will update centroid to mak it maximum to cnetre
    # Step 1: Assign each data point to the nearest centroid

    clusters = [[] for _ in range(k)] # üëà Creates 'k' empty lists inside a list ‚Äî one for each cluster
    for point in X: # For every point in dataset
      # Calculate distance between this point and each centroid
      distances = [euclidean_distance(point, center) for center in centroids]
      nearest_cluster = np.argmin(distances) #finding index of nearest centroid now
      # Add this point to the corresponding cluster's list
      clusters[nearest_cluster].append(point) #like a[i]


      # üì¶ What is clusters here?
# It‚Äôs a list of lists.
# At the start of each iteration, we initialized it like this:
# clusters = [[] for _ in range(k)]
# If k = 3, this makes:
# clusters = [[], [], []]
# Meaning:
# clusters[0] ‚Üí List for points closest to centroid 0
# clusters[1] ‚Üí List for points closest to centroid 1
# clusters[2] ‚Üí List for points closest to centroid 2

# üìå What does cluster_idx hold?
# In this part:
# dists = [distance(point, centroid) for centroid in centroids]  # list of distances
# cluster_idx = np.argmin(dists)  # index of nearest centroid
# dists is a list of distances of this point from each centroid.
# np.argmin(dists) finds the index of the smallest distance ‚Üí meaning the nearest centroid.
# Example:
# If dists = [3.2, 1.5, 7.0]
# then np.argmin(dists) ‚Üí 1 (because 1.5 is smallest)

# üìç Now finally this line:
# clusters[cluster_idx].append(point)
# Meaning:
# The current point gets added to the list at clusters[cluster_idx].
# So if nearest centroid is at index 1, it‚Äôll go into clusters[1].
# Over the loop:
# Each clusters[i] will gradually fill with the points closest to centroid i.




#in avove lien for each itration a for a particular point the min dis ti whch clusetr
#it gets aeed or stored to taht point itsindexin the cluster lsit an we store
#for each same thing so thats why we gave apped..bcz maybe morelets say for cluster 2 tehre willbe 3poist close to it after 5iterations

#clusters = {0: [], 1: [], 2: []}
#point = [2, 3]
# Using append
#clusters[0].append(point)
# Now: clusters = {0: [[2, 3]], 1: [], 2: []}

    # üìå Step 2: Update centroids to mean position of points assigned to them
    for i in range(k):
      points = np.array(clusters[i])
    # Get all points currently assigned to cluster 'idx'
    # clusters[idx] is a list of points ‚Üí convert it to a numpy array for numerical operations

      if len(points) > 0:
        # Check if the cluster has any points assigned to it
        # (If empty, we skip updating to avoid error)

          centroids[i] = points.mean(axis=0)
        # Update the centroid of this cluster
        # Calculate the mean of all points along axis 0 (i.e., feature-wise average)
        # This moves the centroid to the new average position of its assigned points











                        # For every point in dataset
```

```python
#visualizing
colors = ['red', 'blue', 'green']  # Colors for clusters

for i in range(k):
    points = np.array(clusters[i])
    plt.scatter(points[:, 0], points[:, 1], c=colors[i])

```

```python
colors = ['red', 'blue', 'green']  # Colors for clusters

for idx in range(k):
    points = np.array(clusters[idx])
    plt.scatter(points[:, 0], points[:, 1], c=colors[idx], label=f'Cluster {idx+1}')

# üñçÔ∏è Plot the centroids
plt.scatter(centroids[:, 0], centroids[:, 1], marker='*', s=200, c='black', label='Centroids')

plt.title("Final Clusters after K-Means")
plt.legend()
plt.grid(True)
plt.show()
```
