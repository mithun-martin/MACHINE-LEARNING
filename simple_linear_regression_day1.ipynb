```python
#importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import math
```

```python
dataset = pd.read_csv("/content/headbrain.csv")
dataset.head()
```

```python
#so before traoining  model we need to convert the dataframe values in to 2d numpy array
#for that we use df[[attribe]].values
X = dataset[['Head Size(cm^3)']].values  # double brackets to keep it 2D
y = dataset[['Brain Weight(grams)']].values
#[['column']] → returns a DataFrame (2D)
#values → converts it to a NumPy array (still 2D)
```

```python
#visualizing dataset
plt.scatter(X,y)
plt.show()
```

```python
model = LinearRegression()
model.fit(X,y)
```

```python
regression_model_mse = mean_squared_error(X,y)
print('Mean Squared Error:  ', regression_model_mse)
print('root Mean Squared Error:\t',math.sqrt(regression_model_mse))
#So basically — MSE tells you how far your predictions deviate from actual values on average
```

```python
print(model.coef_[0])
"""
model.coef_

- This gives the **coefficient (slope)** of the fitted line.

- It's stored as a NumPy array because scikit-learn’s LinearRegression supports multiple features.

- If you have only one feature or one target variab;le , you access it using [0] because it’s in a 1D array format.

👉 This tells you how much the target variable 'y' changes when the input variable 'X' increases by 1 unit.
"""

print(model.intercept_[0])

"""
model.intercept_

- This gives the **intercept (c)** of the fitted line.

- It's also stored as a NumPy array for consistency, since scikit-learn supports multiple target variables too.

- If you have only one target variable, you access it using [0].

👉 This represents the predicted value of 'y' when 'X' is 0.
In the equation y = m*x + c, this is the value of 'c'.
"""
```

```python
"""
Imagine you trained a simple regression model on height vs weight.

model.coef_ → [2.3]
model.intercept_ → [4.5]

Meaning:

- Every extra **1 unit of height** increases the predicted **weight by 2.3 kg**

- When height is **0 units** (hypothetically), the weight would be **4.5 kg**

So your final regression equation would be:

    Weight = 2.3 × Height + 4.5

Where:
- 2.3 is the **coefficient (slope)** → model.coef_[0]
- 4.5 is the **intercept** → model.intercept_[0]
"""
```

```python
plt.scatter(X,y)
plt.plot(X,model.predict(X), color = 'red')
plt.title('Linear Regression')
plt.xlabel('Head Size')
plt.ylabel('Brain Weight')
```

```python
"""
model.predict() expects the feature(s) (X) as input — the independent variables.

It uses these input features to predict the target variable(s) (y).

You provide the input data (X), and the model returns predicted output values (ŷ).

Passing the actual target values (y) as input to model.predict() is incorrect,
because the model predicts outputs from inputs, not the other way around.
"""


#so basically scatter dots points bw them actually gven b y plt.scatter and plt.
#plot actually draws teh regression line bw x and predicted values of y according to each input value of x
#....which is given by model.predict(x)
```
